```python
import emoji
all_emojis = emoji.get_emoji_unicode_dict("fr")

query = "chienne"

emoji_similarities_to_query={}
for em, str_repr in all_emojis.items():
    if em.replace(':','') not in model_glove_twitter.index_to_key[:50000]:
        continue  # check if we can find next emoji
    sim = model_glove_twitter.similarity(query, em.replace(':',''))
    emoji_similarities_to_query[em] = sim


emoji_similarities_to_query
{k: v for k, v in sorted(emoji_similarities_to_query.items(), key=lambda item: item[1], reverse=True)}
```
